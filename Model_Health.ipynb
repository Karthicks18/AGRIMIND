```python
import pickle
import pandas as pd

# Load the crop model with 'with' statement for safety
model_path = 'models/crop_model.pkl'
with open(model_path, 'rb') as f:
    model = pickle.load(f)

print(f"Model loaded: {type(model)}")  # Sanity check

# Load test data and sample 5 rows
test_data = pd.read_csv('data/Crop_recommendation.csv').sample(5)

# Drop target column if it exists
features = test_data.drop(columns=['label'], errors='ignore')

# Ensure features shape matches model expected input (optional sanity)
print(f"Features shape: {features.shape}")

# Predict only if model has predict method
if hasattr(model, "predict"):
    predictions = model.predict(features)
    print("Test Predictions:", predictions)
else:
    print("Error: Loaded data is not a model with predict method.")
    
print("First row features:\n", features.iloc[0])
```
